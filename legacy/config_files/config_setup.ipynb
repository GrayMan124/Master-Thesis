# A simple script to change the config files

```python
def modify_config(file_path, new_file_path,  changes, add_lines = None):
    """Modify a config file with key=value changes."""
    with open(file_path, "r") as f:
        lines = f.readlines()
    
    # Update lines
    new_lines = []
    for line in lines:
        key = line.split("=")[0].strip()
        if key in changes:
            new_line = f"{key}={changes[key]}\n"
            new_lines.append(new_line)
        else:
            new_lines.append(line)
    if add_lines:
        for line in add_lines:
            new_lines.append(line)
    # Write modified content
    with open(new_file_path, "w") as f:
        f.writelines(new_lines)
```

```python
import re 
for i in range(1,55):
    with open(f'pi_img_config/pi_img_{i}.txt', "r") as f:
        lines = f.readlines()
        
        # Update lines
        new_lines = []
        for line in lines:
            key = line.split("=")[0].strip()
            if key == '--name':
                name = line.split("=")[1].strip()
                # name = name.replace('LAND',f'PI_IMG_{i}')
                name = name.replace('LAND',f'PI_IMG_{i}')
                new_text = re.sub(r'0\.0\d+', '', name)
                new_line = f"{key}={new_text}\n"   
                new_lines.append(new_line)
            else:
                new_lines.append(line)
        # print(new_lines)
        new_file_path = f'pi_img_config/pi_img_{i}.txt'
        with open(new_file_path, "w") as f:
            f.writelines(new_lines)        
```
