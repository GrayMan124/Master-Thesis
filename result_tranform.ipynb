```python
import pandas as pd
import re
```

```python
def parse_log_file(file_path):
    """
    Parse log file with metrics and convert to pandas DataFrame
    
    Parameters:
    file_path (str): Path to the log file
    
    Returns:
    pandas.DataFrame: DataFrame containing parsed metrics
    """
    # Initialize lists to store data
    models = []
    train_losses = []
    val_losses = []
    train_accs = []
    val_accs = []
    
    # Read the file line by line
    with open(file_path, 'r') as file:
        for line in file:
            # Use regex to extract values
            model = re.search(r'Model: (\w+)', line)
            train_loss = re.search(r'train_loss: ([\d.]+)', line)
            val_loss = re.search(r'val_loss: ([\d.]+)', line)
            train_acc = re.search(r'train_acc: ([\d.]+)', line)
            val_acc = re.search(r'val_acc: ([\d.]+)', line)
            
            # Add extracted values to respective lists
            if all([model, train_loss, val_loss, train_acc, val_acc]):
                models.append(model.group(1))
                train_losses.append(float(train_loss.group(1)))
                val_losses.append(float(val_loss.group(1)))
                train_accs.append(float(train_acc.group(1)))
                val_accs.append(float(val_acc.group(1)))
    
    # Create DataFrame
    df = pd.DataFrame({
        'Model': models,
        'Train Loss': train_losses,
        'Validation Loss': val_losses,
        'Train Accuracy': train_accs,
        'Validation Accuracy': val_accs
    })
    
    return df
```

```python
def save_metrics(df, output_path, format='csv'):
    """
    Save the DataFrame to a file
    
    Parameters:
    df (pandas.DataFrame): DataFrame to save
    output_path (str): Path where to save the file
    format (str): Format to save ('csv' or 'excel')
    """
    if format.lower() == 'csv':
        df.to_csv(output_path, index=False)
    elif format.lower() == 'excel':
        df.to_excel(output_path, index=False)
    else:
        raise ValueError("Format must be either 'csv' or 'excel'")
```

```python
input_file = "results_pi_n.txt"
output_file = "metrics_pi_output.csv"  # or "metrics_output.xlsx" for Excel

# Parse the log file and create DataFrame
metrics_df = parse_log_file(input_file)

# Save to CSV or Excel
save_metrics(metrics_df, output_file, format='csv')

# Display first few rows of the DataFrame
print(metrics_df.head())
```
